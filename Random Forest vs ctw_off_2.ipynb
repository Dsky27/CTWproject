```python
import yfinance as yf
import pandas as pd
from ctw3 import CTW
import numpy as np  
import matplotlib.pyplot as plt 
```

# COMPARAISON 

```python
from tqdm import tqdm

# === 1. Fonction d'encodage personnalisée ===
def convert_to_classes(r):
    if r < -0.005:
        return 0  # Strong Negative
    elif -0.005 <= r < 0:
        return 1  # Slightly Negative
    elif 0 <= r < 0.005:
        return 2  # Slightly Positive
    else:
        return 3  # Strong Positive
```

```python
start_date = "2015-01-01"
end_date = "2024-12-31"

sideseq_ticker='DJI'
sideseq_tickertwo='FCHI'
df = yf.download(['^GSPC', f'^{sideseq_ticker}', f'^{sideseq_tickertwo}'], start=start_date, end=end_date)['Close']
df.columns = ['SP500', f'{sideseq_ticker}', f'{sideseq_tickertwo}']
df = df.dropna()
```

```python
#Encodage du SP500 et du VIX
sp500_returns = df['SP500'].pct_change().fillna(0)
vix_returns = df[f'{sideseq_ticker}'].pct_change().fillna(0)
dow_returns = df[f'{sideseq_tickertwo}'].pct_change().fillna(0)

seq = sp500_returns.apply(convert_to_classes).tolist()
sideseq = vix_returns.apply(lambda r: 0 if r < 0 else 1).tolist()

sideseqtwo = dow_returns.apply(lambda r: 0 if r < 0 else 1).tolist()
```

```python
#Alignement des séquences
min_len = min(len(seq), len(sideseq))
seq = seq[:min_len]
sideseq = sideseq[:min_len]
```

```python
#Évaluation de l'accuracy selon la profondeur
accuracies = []
depths = list(range(1, 6))  # profondeur de 1 à 5

for d in tqdm(depths):
    ctw = CTW(depth=d, symbols=4, sidesymbols=2) 
    dist = ctw.predict_sequence(seq, sideseq=sideseq,sideseqtwo=sideseqtwo)
    
    preds = np.argmax(dist, axis=0)
    true_labels = np.array(seq[d:])
    
    acc = np.mean(preds == true_labels)
    accuracies.append(acc)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(depths, accuracies, marker='o')
plt.title("trois sequences")
plt.xlabel("Profondeur de l’arbre (D)")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```
